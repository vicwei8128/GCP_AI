```python
import os

import numpy as np
from PIL import Image
from google.cloud import aiplatform

YOUR_PROJECT = 'My Project'
YOUR_LOCATION = 'us-central1'
YOUR_ENDPOINT = '7687473040107503616'
YOUR_SERVICE = './trees14-moblienetv2-200-acc0.714/Vertex AI.json'
YOUR_PIC = 'tm.jpg'
CAT = ['', '', '', '', '', '', '',
'', '', '', '', '', '', '']

def preprocess_image(f, res=200):
    img = Image.open(f)
    img = img.resize((res, res))
    data = np.array(img.getdata())
    img = data.reshape(1, *img.size, 3)
    img = img.astype('float32')
    img /= 255.
            
    return img

os.environ['GOOGLE_APPLICATION_CREDENTIALS'] = YOUR_SERVICE
aiplatform.init(project=YOUR_PROJECT, location=YOUR_LOCATION)
endpoint = aiplatform.Endpoint(YOUR_ENDPOINT)
img = preprocess_image(YOUR_PIC)  # img shape (1, 200, 200, 3)
prediction = endpoint.predict(instances=img.tolist())  # instances < 1.5M
print(CAT[np.argmax(prediction.predictions)])
```
