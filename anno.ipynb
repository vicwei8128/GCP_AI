```python
import os

from glob import glob

YOUR_BUCKET = 'YOUR_BUCKET'
CAT = ['AE', 'FF', 'PF', 'PP']  # 青楓, 水同木, 大葉山欖, 盾柱木
TRAIN_PATH = 'leaves4-200/train'
TEST_PATH = 'leaves4-200/test'
ANNO = 'leaves4-200-anno.csv'
GCS = f'gs://{YOUR_BUCKET}/leaves4-200'

def anno(of, cats, path, tg, mode='w'):
    TG2DIR = {'training': 'train',
              'test': 'test',
              'validation': 'test'}
    if not tg in TG2DIR:
        return
    d = TG2DIR[tg]
    
    o = open(of, mode)
    for c in cats:
        x = os.path.join(path, c, '*.[jJ][pP][gG]')
        for f in glob(x):
            b = os.path.basename(f)
            blob = '/'.join([GCS, d, c, b])
            o.write(f'{tg}, {blob}, {c}\n')
    o.close()

anno(of=ANNO, cats=CAT, path=TRAIN_PATH, tg='training')
anno(of=ANNO, cats=CAT, path=TEST_PATH, tg='test', mode='a')
```
